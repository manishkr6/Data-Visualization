```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Matplotlib Basics**

```python
x = np.linspace(0, 5, 11)
x
y = x**2
y
```

```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("y axis")
```

## **Subplots Basics**

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.subplot(2, 2, 3)
plt.plot(y, x)
plt.subplot(2, 2, 2)
plt.plot(x**2, x**2)
```

## **Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.Figure()
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.1, 0.1, 1, 1])
axis2.plot(x, y)
```

## **Types of Plots**

```python
plt.scatter(x, y)
```

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

## **Saving Plots**

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("basicplot.png")
```

## **Working with images**

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("images.jpeg")
img
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
cropped_img = img[50:200, 100:300]  # crop: rows 50-200, cols 100-300
plt.imshow(cropped_img)
plt.axis('off')
```
