```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
df = sns.load_dataset("tips")
df.head(5)
```

```python
sns.countplot(x =df['sex'], hue=df['smoker'])
```

```python
sns.barplot(x = df['sex'], y = df['total_bill'])
```

```python
import numpy as np
sns.barplot(x = df['sex'], y = df['tip'], estimator=np.sum)
```

```python
sns.boxplot(x = 'tip', y = 'day', data=df, palette='rainbow')
```

```python
sns.violinplot(x = df['day'], y = df['total_bill'])
```

```python
sns.stripplot(x = df['day'], y = df['total_bill'])
```

```python
sns.swarmplot(x = df['day'], y = df['total_bill'])
```

```python
sns.violinplot(x = df['tip'], y = df['day'])
sns.swarmplot(x = df['tip'], y = df['day'])
```
